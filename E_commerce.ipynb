#### Identify the correlation between product price and the number of times a product has been purchased.

```python
order_items_df.head()
```

```python
import pandas as pd

# Assuming order_items_df is already loaded

# Step 1: Calculate purchase count and average price for each product
product_stats = order_items_df.groupby('product_id').agg({
    'price': 'mean',
    'order_id': 'size'  # Count number of orders which is equivalent to number of purchases
}).reset_index()

# Rename columns for clarity
product_stats = product_stats.rename(columns={'price': 'avg_price', 'order_id': 'purchase_count'})

# Step 2: Calculate correlation between avg_price and purchase_count
correlation = product_stats['avg_price'].corr(product_stats['purchase_count'])

# Display the results
print(f"Correlation between average price and purchase count: {correlation}")
```

#### Calculate the cumulative sales per month for each year.

```python
order_items_df.head()
```

```python
# Step 1: Convert shipping_limit_date to datetime format
order_items_df['shipping_limit_date'] = pd.to_datetime(order_items_df['shipping_limit_date'])

# Step 2: Extract year and month
order_items_df['year'] = order_items_df['shipping_limit_date'].dt.year
order_items_df['month'] = order_items_df['shipping_limit_date'].dt.month

#step 3: calculating the price and freight_values as total_price
order_items_df['total_price']=order_items_df['price']+order_items_df['freight_value']

# Step 4: Group by year and month, then aggregate to calculate cumulative sales
monthly_sales = order_items_df.groupby(['year', 'month'])['total_price'].sum().reset_index()

# Step 5: Calculate cumulative sales per month
monthly_sales['cumulative_sales'] = monthly_sales.groupby('year')['total_price'].cumsum()

# Display the result
print(monthly_sales)
```

#### Calculate the year-over-year growth rate of total sales

```python
order_items_df.head()
```

```python
order_df.head()
```

```python
# Merge dataframes on order_id
merged_df = pd.merge(order_items_df, order_df, on='order_id')

# Convert order_purchase_timestamp to datetime
merged_df['order_purchase_timestamp'] = pd.to_datetime(merged_df['order_purchase_timestamp'])

# Extract year from order_purchase_timestamp
merged_df['year'] = merged_df['order_purchase_timestamp'].dt.year

# calculating the price and freight_values as total_price
order_items_df['total_price']=order_items_df['price']+order_items_df['freight_value']

# Group by year and sum the price
annual_sales = merged_df.groupby('year')['total_price'].sum().reset_index()

# Calculate year-over-year growth rate
annual_sales['YoY_growth'] = annual_sales['total_price'].pct_change() * 100

annual_sales
```

```python
fig, ax1 = plt.subplots(figsize=(12, 8))
ax1.bar(annual_sales['year'], annual_sales['total_price'], color='skyblue', label='Total Sales')
ax1.set_xlabel('Year')
ax1.set_ylabel('Total Sales', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')

# Create a secondary y-axis to plot YoY growth
ax2 = ax1.twinx()
ax2.plot(annual_sales['year'], annual_sales['YoY_growth'], color='orange', marker='o', label='YoY Growth')
ax2.set_ylabel('YoY Growth (%)', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# Adding titles and legends
plt.title('Annual Sales and Year-over-Year Growth Rate')
fig.tight_layout()
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9))

# Show plot
plt.show()
```

 #### Calculate the retention rate of customers, defined as the percentage of customers who make another purchase within 6 months of their first purchase.

```python
order_df.head(10)
```

```python
# Convert order_purchase_timestamp to datetime
order_df['order_purchase_timestamp'] = pd.to_datetime(order_df['order_purchase_timestamp'])

# Identify the first purchase date for each customer
first_purchase_df = order_df.groupby('customer_id')['order_purchase_timestamp'].min().reset_index()
first_purchase_df.columns = ['customer_id', 'first_purchase_date']

# Merge the first purchase date back to the original dataframe
order_df = pd.merge(order_df, first_purchase_df, on='customer_id')

# Calculate 6 months from the first purchase date
order_df['six_months_later'] = order_df['first_purchase_date'] + pd.DateOffset(months=6)

# Determine if there is a subsequent purchase within 6 months
order_df['is_within_6_months'] = (order_df['order_purchase_timestamp'] > order_df['first_purchase_date']) & (order_df['order_purchase_timestamp'] <= order_df['six_months_later'])

# Identify customers who made a subsequent purchase within 6 months
retained_customers = order_df.groupby('customer_id')['is_within_6_months'].any().sum()
total_customers = first_purchase_df['customer_id'].nunique()
retention_rate = (retained_customers / total_customers) * 100

print(f"Customer retention rate: {retention_rate}%")
```

#### Identify the top 3 customers who spent the most money in each year

```python
# Convert 'order_purchase_timestamp' to datetime format
order_df['order_purchase_timestamp'] = pd.to_datetime(order_df['order_purchase_timestamp'])

# Extract the year from 'order_purchase_timestamp'
order_df['year'] = order_df['order_purchase_timestamp'].dt.year

# Merge order_df and order_items_df on order_id
merged_df = pd.merge(order_df, order_items_df, on='order_id')


# Group by customer_id and year, and sum the total spending (price + freight_value)
merged_df['total_spent'] = merged_df['price'] + merged_df['freight_value']
customer_spending = merged_df.groupby(['customer_id', 'year'])['total_spent'].sum().reset_index()

# Find the top 3 customers for each year
top_customers = customer_spending.sort_values(['year', 'total_spent'], ascending=[True, False]).groupby('year').head(3)

top_customers

```

```python
plt.figure(figsize=(13,7))
for year in top_customers['year'].unique():
    yearly_data = top_customers[top_customers['year'] == year]
    plt.bar(yearly_data['customer_id'], yearly_data['total_spent'], label=str(year))

plt.xlabel('Customer ID')
plt.ylabel('Total Spent')
plt.title('Top 3 Customers Who Spent the Most Money Each Year')
plt.legend(title='Year')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
